```python
from utils import show_matrix, matmul
from scipy.sparse import lil_matrix

basis_list = model.basis_list.copy()

X_pd = lil_matrix(X_train.shape)

for i in range(n_basis):
    pattern = matmul(U=basis_list[0][i].T, V=basis_list[1][i])
    X_pd[X_pd > 0] = 1
    X_pd[pattern.astype(bool)] += pattern[pattern.astype(bool)] * 2

    show_matrix(
        [(X_pd, [0, 0], 'pd')], 
        colorbar=True,
        discrete=True,
        center=True, 
        clim=[0, 3], 
        title=f'basis: {i} in [0, {n_basis - 1}]')
```
