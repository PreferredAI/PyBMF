```python
df_netflix_imdb_tmdb = pd.read_pickle("netflix_imdb_tmdb.pickle")

df_netflix_imdb = pd.DataFrame(columns=['netflix_id', 'imdb_id', 'genre', 'cast'])
```

```python
def get_netflix_imdb(i):
    iid = df_netflix_imdb_tmdb.iloc[i]['imdb_id']

    for trial in range(10):
        try:
            if iid is not None:
                movie = ia.get_movie(iid[2:])
                try:
                    genres = movie.data['genres']
                except:
                    genres = []
                    print("[W] Netflix ID {} IMDB ID {} movie.data has no attr 'genres'".format(i, iid))
                cast = movie.get('cast')
                cast = [cast[i].getID() for i in range(len(cast))] if cast is not None else []
            else:
                genres, cast = None, None
            df_netflix_imdb.loc[len(df_netflix_imdb.index)] = [i, iid, genres, cast]
            break
        except Exception as error:
            print("[W] Trial {} Netflix ID {} IMDB ID {} Error {}".format(trial, i, iid, error))


    # return None
```

```python
# fix missing movies during scraping (2/2) repeat several times untill there's no missing movies

from tqdm.contrib.concurrent import thread_map

df_netflix_imdb = pd.DataFrame(columns=['netflix_id', 'imdb_id', 'genres', 'cast'])

r = thread_map(get_netflix_imdb, [i for i in df_netflix_imdb_tmdb.index])
```
