# Shapes

```python
def get_A(m):
    print(m['X_train'].shape, m['U'].shape)

    usize = m['U'].sum(axis=0)
    vsize = m['V'].sum(axis=0)

    usize = np.array(usize).flatten()
    vsize = np.array(vsize).flatten()

    k = len(usize)
    umax = usize.max()
    vmax = vsize.max()
    
    A = np.zeros((int(umax), int(vmax)))

    for i in range(k):
        if usize[i] == 0 or vsize[i] == 0:
            continue
        A[int(usize[i]-1), int(vsize[i]-1)] += 1

    # print(A)
    return A
```

```python
def get_A_hyper(m):
    # if 'U_100' in m['logs'].keys():
    #     usize = m['logs']['U_100'].sum(axis=0)
    #     vsize = m['logs']['V_100'].sum(axis=0)
    # else:
    if len(m['logs']['U']) > 100:
        usize = m['logs']['U'][-100].sum(axis=0)
        vsize = m['logs']['V'][-100].sum(axis=0)
    else:
        usize = m['logs']['U'][0].sum(axis=0)
        vsize = m['logs']['V'][0].sum(axis=0)

    usize = np.array(usize).flatten()
    vsize = np.array(vsize).flatten()

    k = len(usize)
    umax = usize.max()
    vmax = vsize.max()
    
    A = np.zeros((int(umax), int(vmax)))

    for i in range(k):
        if usize[i] == 0 or vsize[i] == 0:
            continue
        A[int(usize[i]-1), int(vsize[i]-1)] += 1

    # print(A)
    return A
```

```python
def get_FP_FN_hyper(m, to_print=False):
    # if hyper:
    #     if 'U_100' in m['logs'].keys():
    #         U, V = m['logs']['U_100'], m['logs']['V_100']
    #     else:
    #         if len(m['logs']['U']) > 100:
    #             U, V = m['logs']['U'][-100], m['logs']['V'][-100]
    #         else:
    #             U, V = m['logs']['U'][0], m['logs']['V'][0]
    # else:
    # U, V = m['U'], m['V']
        
    X_train = m['X_train']

    fp_list, fn_list, i_list = [], [], []

    # stop_flag = False
    # stop_sign = [150, 243]

    for i in range(min(100, len(m['logs']['U']))):
        print(i, min(100, len(m['logs']['U'])))
        U, V = m['logs']['U'][-(i+1)], m['logs']['V'][-(i+1)]

        X_pd = matmul(U, V.T, boolean=True, sparse=True)

        fp = FP(gt=X_train, pd=X_pd)
        fn = FN(gt=X_train, pd=X_pd)

        if to_print: 
            print(fp, fn)

        # if fp == stop_sign[0] and fn == stop_sign[1]:
        #     break

        fp_list.append(fp / X_train.sum())
        fn_list.append(fn / X_train.sum())
        i_list.append(i)

    # make it 100
    if len(i_list) < 100:
        i_list = i_list + [i_list[-1], 100]
        fp_list = fp_list + [0, 0]
        fn_list = fn_list + [0, 0]

    return fp_list, fn_list, i_list


##############################################################################


def get_FP_FN(m, stop_sign=[-1, -1],to_print=False):

    U, V = m['U'], m['V']
        
    X_train = m['X_train']

    fp_list, fn_list, i_list = [], [], []

    # stop_flag = False
    # stop_sign = [150, 243]

    for i in range(max(100, U.shape[1])):
        X_pd = matmul(U[:, :i+1], V[:, :i+1].T, boolean=True, sparse=True)

        fp = FP(gt=X_train, pd=X_pd)
        fn = FN(gt=X_train, pd=X_pd)

        if to_print: 
            print(fp, fn)

        if fp == stop_sign[0] and fn == stop_sign[1]:
            break

        fp_list.append(fp / X_train.sum())
        fn_list.append(fn / X_train.sum())
        i_list.append(i)

    # make it 100
    if len(i_list) < 100:
        i_list = i_list + [i_list[-1], 100]
        fp_list = fp_list + [0, 0]
        fn_list = fn_list + [0, 0]

    return fp_list, fn_list, i_list


def plot_FP_FN(fp_list, fn_list, i_list, title):
    import matplotlib.pyplot as plt
    import numpy as np
    import matplotlib.ticker as mticker

    DICT = {
        'FP/P': fp_list,
        'FN/P': fn_list
    }
    color_map = ["r", "orange"]

    fig, ax = plt.subplots()
    ax.stackplot(i_list, DICT.values(),
                labels=DICT.keys(), alpha=0.5, colors = color_map)
    ax.legend(loc='upper right')

    ax.set_title(title)
    ax.set_xlabel('k')
    ax.set_ylabel('error')
    # add tick at every 200 million people
    # ax.yaxis.set_minor_locator(mticker.MultipleLocator(.2))
    fig.savefig(title + ".pdf", transparent=True)


    fig.show()
```
