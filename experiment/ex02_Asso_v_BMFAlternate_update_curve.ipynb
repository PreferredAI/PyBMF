```python
import matplotlib.pyplot as plt
import pandas as pd
import pickle

# df = model.logs['scores'].copy()

with open('../saved_models/24-05-21_01-43-10_BMFAlternate.pickle', 'rb') as f:
    m = pickle.load(f)
df = m['logs']['scores'].copy()
df = df.drop(df.columns[0], axis=1)
df
```

```python
# worst candidates

d = df.iloc[13].tolist()
worst_candidates = [i for i, value in enumerate(d) if value == min(d)] # 0
```

```python
# worst init

# sorted(set(d))

d = df.iloc[0].tolist()
worst_inits = [i for i, value in enumerate(d) if value == 0.5] #53

worst_inits
```

```python
# group 1

d = df.iloc[13].tolist()
group_1 = [index for index, value in enumerate(d) if 1750 > value > 1500]
group_2 = [index for index, value in enumerate(d) if 1250 > value > 1000]
group_3 = [index for index, value in enumerate(d) if 500 > value > 250]
```

```python
group_3
```

```python
cmap_1 = plt.colormaps['autumn']
cmap_2 = plt.colormaps['winter']
cmap_3 = plt.colormaps['summer']
```

```python
# # greatest improvement

# greatest_improvement = ((1700 > df.iloc[9]) & (df.iloc[9] > 1500)).tolist().index(True)
# greatest_improvement
```

```python
# best candidate

d = df.iloc[0].tolist()
best_candidate = d.index(max(d))
best_init_score = max(d)
```

```python
import numpy as np


df = m['logs']['scores'].copy()
# df = model.logs['scores'].copy()
df = df.drop(df.columns[0], axis=1)
df = df.head(12)

fig, ax = plt.subplots(figsize=(8, 6), dpi=100)

ax.axhline(best_init_score, color='red', linestyle='dashdot', label='candidate score given by Asso')


# m = 1e10
idx = 0
for i, column in enumerate(df.columns):
    if i % 10 != 0 or i in worst_candidates:
        continue
    
    if i in group_1:
        color = cmap_1(np.random.uniform(.2, .6))
        alpha = 0.1
    elif i in group_2:
        color = cmap_2(np.random.uniform(.2, .6))
        alpha = 0.1
    elif i in group_3:
        color = cmap_3(np.random.uniform(.2, .6))
        alpha = 0.5

    # ax.scatter(df.index, df[column], color=color, alpha=0.5)
    ax.plot(df.index, df[column], linestyle='--', marker='o', markersize=6, color=color, alpha=alpha)


for i, column in enumerate(df.columns):
    if i == best_candidate:
        # ax.scatter(df.index, df[column], color='orange')
        ax.plot(df.index, df[column], linestyle='-', marker='o', markersize=6, label='candidate with highest score', color='red', alpha=1)
        continue

    if i == worst_inits[1]:
        ax.scatter(df.index, df[column], color='red')
        ax.plot(df.index, df[column], linestyle='-', marker='o', markersize=6, label='candidate with lowest initial score', color='orange', alpha=1)
        continue


ax.set_xlabel('steps')
ax.set_ylabel('score')

handles, labels = plt.gca().get_legend_handles_labels()
order = [0,2,1]
plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order])

# ax.legend()
plt.grid(ls=':')

fig1 = plt.gcf()

plt.show()
plt.draw()

fig1.savefig('ex02.pdf')
```
