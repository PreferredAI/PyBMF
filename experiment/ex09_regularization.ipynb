```python
import matplotlib.pyplot as plt
import numpy as np

def bowl(x):
    return (x*(x-1)) **2

def primp(x):
    return np.minimum(np.abs(x), np.abs(x-1))

def elb(x):
    return np.minimum(np.abs(x) + x ** 2, np.abs(x-1) + (x-1) ** 2)

x = np.linspace(-1, 2, 1000)
xx = np.linspace(0, 1, 10000)

plt.figure(figsize=(10, 8))
plt.plot(x, bowl(x), '-', color='skyblue', linewidth=2, label='BinaryMF-Penalty, Mexican hat')
plt.plot(xx, primp(xx), '-', color='purple', linewidth=2, label='PRIMP, $\Lambda$')
plt.plot(x, elb(x), '-', color='navy', linewidth=2, label='ELBMF, $R$')

plt.xlabel('x', fontsize=14)
plt.ylabel('R(x)', fontsize=14)
# plt.title('Regularizer', fontsize=16)
plt.legend(fontsize=12)
plt.ylim(0, 1)
plt.xlim(-1, 2)

# Add arrows for PRIMP
plt.annotate('', xy=(0, 1), xytext=(0, 0), arrowprops=dict(arrowstyle='->', color='purple', lw=1.5))
plt.annotate('', xy=(1, 1), xytext=(1, 0), arrowprops=dict(arrowstyle='->', color='purple', lw=1.5))

plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def bowl(x):
    return (x*(x-1)) **2

def primp(x):
    return np.minimum(np.abs(x), np.abs(x-1))

def elb(x):
    return np.minimum(np.abs(x) + x ** 2, np.abs(x-1) + (x-1) ** 2)

x = np.linspace(-1, 2, 1000)
xx = np.linspace(0, 1, 10000)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(6, 6), sharex=True)

# BinaryMF-Penalty plot
ax1.plot(x, bowl(x), '-', color='skyblue', linewidth=3)
ax1.set_xlabel('x', fontsize=14)
ax1.set_ylabel('f(x)', fontsize=14)
ax1.set_title('BinaryMF-Penalty, the Mexican hat', fontsize=16)
ax1.set_ylim(0, 1)
ax1.set_xlim(-1, 2)
ax1.grid(True, linestyle='--', alpha=0.7)

# PRIMP plot
ax2.plot(xx, primp(xx), '-', color='purple', linewidth=3)
ax2.set_xlabel('x', fontsize=14)
ax2.set_ylabel('f(x)', fontsize=14)
ax2.set_title('PRIMP, the $\Lambda$ function', fontsize=16)
ax2.set_ylim(0, 1)
ax2.set_xlim(-1, 2)
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.annotate('', xy=(0, 1), xytext=(0, 0), arrowprops=dict(arrowstyle='->', color='purple', lw=3))
ax2.annotate('', xy=(1, 1), xytext=(1, 0), arrowprops=dict(arrowstyle='->', color='purple', lw=3))

# ELBMF plot
ax3.plot(x, elb(x), '-', color='navy', linewidth=3)
ax3.set_xlabel('x', fontsize=14)
ax3.set_ylabel('f(x)', fontsize=14)
ax3.set_title('ELBMF, the $R$ function', fontsize=16)
ax3.set_ylim(0, 1)
ax3.set_xlim(-1, 2)
ax3.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig("img_regularizers.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
