```python
def get_FP_FN_hyper(m, to_print=False):
    # if hyper:
    #     if 'U_100' in m['logs'].keys():
    #         U, V = m['logs']['U_100'], m['logs']['V_100']
    #     else:
    #         if len(m['logs']['U']) > 100:
    #             U, V = m['logs']['U'][-100], m['logs']['V'][-100]
    #         else:
    #             U, V = m['logs']['U'][0], m['logs']['V'][0]
    # else:
    # U, V = m['U'], m['V']
        
    X_train = m['X_train']

    fp_list, fn_list, i_list = [], [], []

    # stop_flag = False
    # stop_sign = [150, 243]

    for i in range(min(100, len(m['logs']['U']))):
        # print(i, min(100, len(m['logs']['U'])))
        U, V = m['logs']['U'][-(i+1)], m['logs']['V'][-(i+1)]

        X_pd = matmul(U, V.T, boolean=True, sparse=True)

        fp = FP(gt=X_train, pd=X_pd)
        fn = FN(gt=X_train, pd=X_pd)

        if to_print: 
            print(fp, fn)

        # if fp == stop_sign[0] and fn == stop_sign[1]:
        #     break

        fp_list.append(fp / (X_train.shape[0] * X_train.shape[1]))
        fn_list.append(fn / (X_train.shape[0] * X_train.shape[1]))
        i_list.append(i)

    # make it 100
    if len(i_list) < 100:
        i_list = i_list + [i_list[-1], 100]
        fp_list = fp_list + [0, 0]
        fn_list = fn_list + [0, 0]

    return fp_list, fn_list, i_list


##############################################################################


def get_FP_FN(m, stop_k=-1 ,to_print=False):

    U, V = m['U'], m['V']
        
    X_train = m['X_train']

    fp_list, fn_list, i_list = [], [], []

    # stop_flag = False
    # stop_sign = [150, 243]

    for i in range(max(100, U.shape[1])):
        X_pd = matmul(U[:, :i+1], V[:, :i+1].T, boolean=True, sparse=True)

        fp = FP(gt=X_train, pd=X_pd)
        fn = FN(gt=X_train, pd=X_pd)

        if to_print: 
            print(fp, fn)

        if i == stop_k:
            break

        fp_list.append(fp / (X_train.shape[0] * X_train.shape[1]))
        fn_list.append(fn / (X_train.shape[0] * X_train.shape[1]))
        i_list.append(i)

    # make it 100
    if len(i_list) < 100:
        i_list = i_list + [i_list[-1], 100]
        fp_list = fp_list + [0, 0]
        fn_list = fn_list + [0, 0]

    return fp_list, fn_list, i_list


def plot_FP_FN(fp_list, fn_list, i_list, title):
    import matplotlib.pyplot as plt
    import numpy as np
    import matplotlib.ticker as mticker

    DICT = {
        'FP/P': fp_list,
        'FN/P': fn_list
    }
    color_map = ["r", "orange"]

    fig, ax = plt.subplots(figsize=[5, 4])
    ax.stackplot(i_list, DICT.values(),
                labels=DICT.keys(), alpha=0.5, colors = color_map)
    ax.legend(loc='upper right')

    ax.set_title(title)
    ax.set_xlabel('k')
    ax.set_ylabel('Error rate')
    # add tick at every 200 million people
    # ax.yaxis.set_minor_locator(mticker.MultipleLocator(.2))
    fig.savefig("img_errors_" + title + ".pdf", transparent=True)


    fig.show()
```

```python
i = 0

m1 = models_asso[i]
m2 = models_panda[i]
m3 = models_hyper[i]
m4 = models_mebf[i]

fp_list, fn_list, i_list = get_FP_FN(m1, stop_k = 52)
plot_FP_FN(fp_list, fn_list, i_list, datasets[i] + " (Asso)")

fp_list, fn_list, i_list = get_FP_FN(m2)
plot_FP_FN(fp_list, fn_list, i_list, datasets[i] + " (Panda)")

fp_list, fn_list, i_list = get_FP_FN_hyper(m3)
plot_FP_FN(fp_list, fn_list, i_list, datasets[i] + " (Hyper)")

fp_list, fn_list, i_list = get_FP_FN(m4)
plot_FP_FN(fp_list, fn_list, i_list, datasets[i] + " (MEBF)")
```
