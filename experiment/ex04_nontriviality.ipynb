```python
def get_FP_FN_hyper(m, to_print=False):

    # if 'U_100' in m['logs'].keys():
    #     U, V = m['logs']['U_100'], m['logs']['V_100']
    # else:
    #     if len(m['logs']['U']) > 100:
    #         U, V = m['logs']['U'][-100], m['logs']['V'][-100]
    #     else:
    #         U, V = m['logs']['U'][0], m['logs']['V'][0]

        
    X_train = m['X_train']

    nontriv_list, overlap_list, triv_list, i_list = [], [], [], []
    triv_count = []

    # stop_flag = False
    # stop_sign = [150, 243]

    for i in range(min(100, len(m['logs']['U']))):
        # print(i, min(100, len(m['logs']['U'])))
        U, V = m['logs']['U'][-(i+1)], m['logs']['V'][-(i+1)]

        # count nontrivial patterns
        triv_u = np.array(U.sum(axis=0) == 1).flatten()
        triv_v = np.array(V.sum(axis=0) == 1).flatten()
        triv_id = np.logical_or(triv_u, triv_v)
        nontriv_id = np.logical_not(triv_id)

        triv_count.append(triv_id.sum())
        
        # measure nontrivial areas
        nontriv_area = matmul(U[:, nontriv_id], V[:, nontriv_id].T, boolean=True, sparse=True) if nontriv_id.sum() > 0 else lil_matrix(X_train.shape)

        triv_area = matmul(U[:, triv_id], V[:, triv_id].T, boolean=True, sparse=True) if triv_id.sum() > 0 else lil_matrix(X_train.shape)

        ol_area = multiply(triv_area, nontriv_area, boolean=True, sparse=True)

        nt, ol, t = nontriv_area.sum() - ol_area.sum(), ol_area.sum(), triv_area.sum()

        # print(i, nt, ol, t)

        nontriv_list.append(nt / (X_train.shape[0] * X_train.shape[1]))
        overlap_list.append(ol / (X_train.shape[0] * X_train.shape[1]))
        triv_list.append(t / (X_train.shape[0] * X_train.shape[1]))

        i_list.append(i)

    # # make it 100
    # if len(i_list) < 100:
    #     i_list = i_list + [i_list[-1], 100]
    #     fp_list = fp_list + [0, 0]
    #     fn_list = fn_list + [0, 0]

    return nontriv_list, overlap_list, triv_list, i_list, triv_count


##############################################################################


def get_FP_FN(m, stop_k=-1,to_print=False):

    U, V = m['U'], m['V']
        
    X_train = m['X_train']

    nontriv_list, overlap_list, triv_list, i_list = [], [], [], []
    triv_count = []
    # stop_flag = False
    # stop_sign = [150, 243]

    for i in range(max(100, U.shape[1])):
        # X_pd = matmul(U[:, :i+1], V[:, :i+1].T, boolean=True, sparse=True)

        U, V = m['U'][:, :i+1], m['V'][:, :i+1]

        # count nontrivial patterns
        triv_u = np.array(U.sum(axis=0) == 1).flatten()
        triv_v = np.array(V.sum(axis=0) == 1).flatten()
        triv_id = np.logical_or(triv_u, triv_v)
        nontriv_id = np.logical_not(triv_id)

        triv_count.append(triv_id.sum())
        
        # measure nontrivial areas
        nontriv_area = matmul(U[:, nontriv_id], V[:, nontriv_id].T, boolean=True, sparse=True) if nontriv_id.sum() > 0 else lil_matrix(X_train.shape)

        triv_area = matmul(U[:, triv_id], V[:, triv_id].T, boolean=True, sparse=True) if triv_id.sum() > 0 else lil_matrix(X_train.shape)

        ol_area = multiply(triv_area, nontriv_area, boolean=True, sparse=True)

        nt, ol, t = nontriv_area.sum() - ol_area.sum(), ol_area.sum(), triv_area.sum()

        # print(i, nt, ol, t)

        if i == stop_k:
            break

        nontriv_list.append(nt / (X_train.shape[0] * X_train.shape[1]))
        overlap_list.append(ol / (X_train.shape[0] * X_train.shape[1]))
        triv_list.append(t / (X_train.shape[0] * X_train.shape[1]))

        i_list.append(i)

    # make it 100
    if len(i_list) < 100:
        i_list = i_list + [i_list[-1], 100]
        nontriv_list = nontriv_list + [0, 0]
        overlap_list = overlap_list + [0, 0]
        triv_list = triv_list + [0, 0]

    return nontriv_list, overlap_list, triv_list, i_list, triv_count


def plot_FP_FN(nt_list, ol_list, t_list, i_list, triv_count, title):
    import matplotlib.pyplot as plt
    import numpy as np
    import matplotlib.ticker as mticker

    DICT = {
        'Trivial': t_list,
        'Overlapping': ol_list, 
        'Non-trivial': nt_list,
    }
    color_map = ["r", "gold", 'cyan']

    fig, ax = plt.subplots(figsize=[5, 4])
    ax.stackplot(i_list, DICT.values(),
                labels=DICT.keys(), alpha=0.35, colors = color_map)
    ax.set_ylabel('Ratio in whole matrix')
    ax.legend(loc='upper left')

    # curve
    ax2 = ax.twinx()
    ax2.step(i_list[:len(triv_count)], triv_count, color='red', label='N trivial', alpha=0.7)
    ax2.set_ylabel('Number of trivial patterns')
    ax2.legend(loc='upper right')
    ax2.set_ylim(0, 100)

    ax.set_title(title)
    ax.set_xlabel('k')
    # ax.set_ylabel('error')
    # add tick at every 200 million people
    # ax.yaxis.set_minor_locator(mticker.MultipleLocator(.2))
    fig.savefig("img_triviality_" + title + ".pdf", transparent=True)


    fig.show()
```
